```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

Setting system paramters:

```python
V0 = 1e-6
f0 = 0.6
a = 0.015
b = 0.02
eta = 4.5e6
L = 1e-1
sn = 50e6
Vinit = 1e-9
Vp = 1e-8
k = 500000
yeild_point_init = 10
final_time = 500 * (365 * 24 * 3600)
```

```python
kcrit = sn * b / L
print('k/kcrit =', k / kcrit)
```

```python
os.system(f"./app {V0} {f0} {a} {b} {eta} {L} {sn} {Vinit} {Vp} {k} {yeild_point_init} {final_time} out.txt")
print(f"./app {V0} {f0} {a} {b} {eta} {L} {sn} {Vinit} {Vp} {k} {yeild_point_init} {final_time} out.txt")
```

```python
df = pd.read_csv('./out.txt')
df['t'] = df.t / (365 * 24 * 3600)
```

```python
fig, axs = plt.subplots(2, 2, sharex=True, figsize=(12, 5))
ax = axs[0, 0]
ax.plot(df.t, df.D)
ax.set_ylabel('displacement [m]')


ax = axs[0, 1]
ax.plot(df.t, df.V)
ax.set_yscale('log')
ax.yaxis.tick_right()
ax.set_ylabel('velocity [m/s]')
ax.yaxis.set_label_position("right")
ax.set_ylim(bottom=1e-17)


ax = axs[1, 0]
ax.plot(df.t, df.tau * 1e-6)
ax.set_ylabel('tau [MPa]')
ax.set_xlabel('time [years]')


ax = axs[1, 1]
plt.plot(df.t, df.psi)
ax.yaxis.tick_right()
ax.set_ylabel('state')
ax.yaxis.set_label_position("right")
plt.savefig('output.png')
ax.set_xlabel('time [years]')
```

```python
plt.scatter(np.arange(len(dt)), dt, s=4)
plt.yscale('log')
```

```python
def psi_init(tau, eta, Vinit, a, sn, V0):
    s = np.sinh((tau - eta * Vinit) / (a * sn))
    l = np.log((2.0 * V0 / Vinit) * s)
    return a * l
```

```python
tau = np.logspace(0, 7, 1000)
plt.figure()
plt.plot(tau, psi_init(tau, eta, Vinit, a, sn, V0))
plt.xscale('log')
```
